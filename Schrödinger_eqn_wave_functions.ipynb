```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def ground_state_wave_function(x, t, m, omega, hbar):
    prefactor = (m * omega / (np.pi * hbar)) ** 0.25
    exponent = -0.5 * (m * omega / hbar) * x ** 2 - 1j * 0.5 * omega * t
    return prefactor * np.exp(exponent)
```

```python
def first_excited_state_wave_function(x, t, m, omega, hbar):
    prefactor = (m * omega / (np.pi * hbar)) ** 0.25 * np.sqrt(2)
    exponent = -0.5 * (m * omega / hbar) * x ** 2 - 1j * 1.5 * omega * t
    return prefactor * x * np.exp(exponent)
```

```python
def second_excited_state_wave_function(x, t, m, omega, hbar):
    prefactor = (m * omega / (np.pi * hbar)) ** 0.25 * (1 / np.sqrt(2))
    term = 2 * (m * omega / hbar) * x ** 2 - 1
    exponent = -0.5 * (m * omega / hbar) * x ** 2 - 1j * 2.5 * omega * t
    return prefactor * term * np.exp(exponent)
```

```python
# Parameters
x_values = np.linspace(-5, 5, 1000)
t = 0.0
m = 1.0
omega = 1.0
hbar = 1.0

# Calculate wave functions
psi_0 = ground_state_wave_function(x_values, t, m, omega, hbar)
psi_1 = first_excited_state_wave_function(x_values, t, m, omega, hbar)
psi_2 = second_excited_state_wave_function(x_values, t, m, omega, hbar)
```

```python
# Plot ground state wave function
plt.figure(figsize=(10, 6))
plt.plot(x_values, np.real(psi_0), label="Ground State Re", color='b')
plt.plot(x_values, np.imag(psi_0), label="Ground State Im", linestyle='dashed', color='b')
plt.title("Ground State Wave Function")
plt.xlabel("Position")
plt.ylabel("Wave Function Value")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Plot first excited state wave function
plt.figure(figsize=(10, 6))
plt.plot(x_values, np.real(psi_1), label="First Excited State Re", color='r')
plt.plot(x_values, np.imag(psi_1), label="First Excited State Im", linestyle='dashed', color='r')
plt.title("First Excited State Wave Function")
plt.xlabel("Position")
plt.ylabel("Wave Function Value")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Plot second excited state wave function
plt.figure(figsize=(10, 6))
plt.plot(x_values, np.real(psi_2), label="Second Excited State Re", color='g')
plt.plot(x_values, np.imag(psi_2), label="Second Excited State Im", linestyle='dashed', color='g')
plt.title("Second Excited State Wave Function")
plt.xlabel("Position")
plt.ylabel("Wave Function Value")
plt.legend()
plt.grid(True)
plt.show()
```
